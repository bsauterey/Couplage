## Compute the dynamics of the ecosystem

```python
t0 = libtime.time()
### Here is a description of the parameters of interest of the model:
"""
Cell:
- rc     : cell radius
- Vc     : cell volume
- Qc     : structural biomass of the cell (depends on rc)

Reproduction and death:
- gmax   : maximum of g, the division rate (of a sigmo√Ød function of the log-internal state of the cell)
- thresh : inflexion point of g, the rate, at which g=gmax/2 (depends on rc)
- slope  : slope in of g at g(thresh)
- kd     : maximum metabolism related decay rate (when metabolism is insufficient to fuel cell maintenance; see below)
- mort   : intrinsic metabolic rate, can be seen as a senescence term or an accidental death term. 

Metabolism:
- qmax   : maximum metabolic rate (Michaelis-Menten; depends on rc)
- ks     : half-saturation constant of the metabolism (Michaelis-Menten)
- mg     : minimum metabolic rate to fulfill the energetic requirement for cell maintainance (depends on rc)

Environment:
- QH, QC, QN, QG : inward/outward fluxes constants. From SBL (diffusion and solubilisation) for H2 (H), CO2 (C), and CH4 (G).
           From hydrothermal vents for NH4 (N). 
- Hinf, Cinf, Ninf, Ginf : abiotic equilibrium values (fixed for the moment assuming no feedback onto 
           atmospheric composition). Equals to 0 for Xo (the dead biomass)
- Still need a term describing the externalization of Xo (dead biomass). For the moment it is described by a constant 
  (burrial?).
"""

### Default parametrization
## t is time (...):
bio_tmax   = 360/12                     
dt         = 0.0025                     
t          = 0
Hinf       = 7.6e-4   * 1 * 7.8e-4    # mol.L-1           Kharecha
#Hinf       =  1e-5                   # mol.L-1           Kharecha
Cinf       = 2.5*1e-3 * 1 * 1e-3      # mol.L-1           Kharecha
Ninf       = 1e-8                     # mol.L-1           arbitrary
Ginf       = 1e-20    * 1 * 1.4e-3    # mol.L-1           Kharecha

starters = [rc,Vc,Qc,ks,qmax,mg,kd,mort,thresh,slope,gmax]
Env = [Hinf,Cinf,Ninf,Ginf,QH,QC,QN,QG]
init = [Hinf,Cinf,Ninf,1E-10,0,1e7,thresh]
NCT,XT,HT,CT,NT,GT,XoT,D,time,NPPT = Run_Profile(init,starters,Env,tmax=bio_tmax,T=TS,dt = dt)
NCT  = np.array(NCT)
XT   = np.array(XT)
HT   = np.array(HT)
CT   = np.array(CT)
NT   = np.array(NT)
GT   = np.array(GT)
XoT  = np.array(XoT)
D    = np.array(D)
time = np.array(time)

FH   = QH*(Hinf-np.mean(HT[-1000:]))
FG   = QG*(Ginf-np.mean(GT[-1000:]))

t1 = libtime.time()
total = t1-t0
floor(total)
```

## Compute the variation of the atmospheric composition

```python
t      = 0
prompt = 0
dt     = 0.00000000000001    # year
tmax   = 1e11

#Environmental constants
Av     = 6e23
p      = 1e5      # Pa     
R      = 8.31     # J.(K.mol)-1
Toc    = 283      # K
S      = 5.1e18   # cm2
g      = 9.81     #
Mair   = 28.16e-3 # kg.mol-1
ntot   = (p*S)/(g*Mair)

#Initial atompheric composition
Hatm   = ntot*7.634e-4     #mol
Gatm   = ntot*1e-20      #mol
COatm  = ntot*1e-20      #mol
Catm   = ntot*2.5*1e-3   #mol

#Photolysis of CH4
a      = 1.438e15 #yr-1
b      = 2.0291
Phot   = -a*(Gatm/ntot)**(1/b) 

#CO conversion
Vco    = 6.3 #yr-1
Conv   = Vco*COatm*p / (ntot*R*Toc)

#Volcanism
Volc   = 1e-5

#Escape
Esc    = -13.1e-4*(Hatm+2*Gatm)/(ntot) 

HatmT   = []     
GatmT   = []     
COatmT  = []     
CatmT   = []     
TotatmT = []     
time    = []     

while t < tmax:
    #if(prompt > tmax/1000000):
    HatmT.append(Hatm)
    GatmT.append(Gatm)
    COatmT.append(COatm)
    time.append(t)
    print("Progress => {:2.1%} ; dt => {:2.1}".format(t/tmax,dt), end="\r")
    prompt=0
        
    #dHatm  = -2*Phot + S * (FH + Esc + Volc + Conv)
    dHatm  = S * (Esc + Volc)
    #dGatm  = Phot + S * (FG)
    dGatm  = 0
    #dCOatm = -2*Phot - S * Conv
    dCOatm = 0
    
    vec = np.array([Hatm,Gatm,COatm])
    der_vec = np.array([dHatm,dGatm,dCOatm])
    
    dt = min(abs(vec[np.where(der_vec != 0)[0]]/der_vec[np.where(der_vec != 0)[0]]/1000))
    #dt = tmax/10000000
    Hatm   = max(1e-40,Hatm + dHatm*dt)
    Gatm   = max(1e-40,Gatm + dGatm*dt)
    COatm  = max(1e-40,COatm + dCOatm*dt)
    
    Phot   = -a*(Gatm/ntot)**(1/b) 
    Conv   = Vco*COatm*p / (ntot*R*Toc)
    Esc    = -13.1e-4*(Hatm+2*Gatm)/(ntot) 

    t      = t + dt
    prompt = prompt + dt

    
HatmT   = np.array(HatmT)   
GatmT   = np.array(GatmT)     
COatmT  = np.array(COatmT)     
CatmT   = np.array(CatmT)     
TotatmT = np.array(TotatmT) 

```

```python
fig, ax = plt.subplots(1,3,sharex=False)

ax[0].semilogy(time,HatmT)
#ax[0].set_ylim([1E18,1E24])
ax[1].semilogy(time,GatmT)
ax[2].semilogy(time,COatmT)
#ax[2].set_ylim([1E-7,1E-1])

fig.subplots_adjust(hspace=0.5,wspace=0.35)
fig.set_figwidth(15)
fig.set_figheight(3)
#fig.add_subplot(111, frameon=False)

plt.show()
```
